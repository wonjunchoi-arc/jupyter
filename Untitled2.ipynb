```python
import pandas as pd
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0,100)
```

```python
df=pd.DataFrame({"국어":[score()for i in range(10)],"영어":[score()for i in range(10)],"수학":[score()for i in range(10)],"사회":[score()for i in range(10)],}, index=[id() for i in range(10)])
```
